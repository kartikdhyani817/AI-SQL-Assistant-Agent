```python
import google.generativeai as genai

API_KEY = "AQ.Ab8RN6KIcWb766pXGyAGg3ISNPsGDJfdWCJDGivZ6zNSgPC5VA"

genai.configure(api_key=API_KEY)

model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
schema = """
Table: Customers

customer_id
customer_name
city

Table: Orders

order_id
customer_id
product
amount
order_date
"""

print(schema)
```

```python

user_question = """
Show all customers who spent more than 5000.
"""
```

```python
prompt = f"""
You are an expert SQL Developer.

Database Schema:

{schema}

User Request:

{user_question}

Generate only the SQL query.
"""
```

```python
response = model.generate_content(prompt)

generated_sql = response.text

print(generated_sql)
```

```python
with open(
    "generated_sql.txt",
    "w",
    encoding="utf-8"
) as file:

    file.write(generated_sql)

print("SQL Saved")
```

```python
explanation_prompt = f"""
You are an SQL instructor.

Explain the following SQL query in simple language.

SQL Query:

{generated_sql}

Explain:
1. What the query does
2. Tables used
3. Conditions used
4. Expected output
"""
```

```python
explanation_response = model.generate_content(
    explanation_prompt
)

sql_explanation = explanation_response.text

print(sql_explanation)
```

```python
with open(
    "sql_explanation.txt",
    "w",
    encoding="utf-8"
) as file:

    file.write(sql_explanation)

print("SQL Explanation Saved")
```
